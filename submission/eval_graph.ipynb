```python
%pwd
%cd "D:\文档\NYU Classes\2023 Fall\ECE-GY 6143 Machine Learning\Lab 11\submission"
%cd "old model (mbnv2)/evaluation_metric"

old_hpa_performance = pd.read_csv('load_output_hpa.csv')
old_hpa_performance['      Date & Time'] = pd.to_datetime(old_hpa_performance['      Date & Time'])
old_lb_performance = pd.read_csv('load_output_lb.csv')
old_lb_performance['      Date & Time'] = pd.to_datetime(old_lb_performance['      Date & Time'])
old_hpa_usage = pd.read_csv('resource_usage_hpa.csv')
old_lb_usage = pd.read_csv('resource_usage_lb.csv')

%cd "D:\文档\NYU Classes\2023 Fall\ECE-GY 6143 Machine Learning\Lab 11\submission"
%cd "MobileNet/evaluation_metric"
new_hpa_performance = pd.read_csv('load_output_hpa.csv')
new_hpa_performance['      Date & Time'] = pd.to_datetime(new_hpa_performance['      Date & Time'])
new_lb_performance = pd.read_csv('load_output_lb.csv')
new_lb_performance['      Date & Time'] = pd.to_datetime(new_lb_performance['      Date & Time'])
new_hpa_usage = pd.read_csv('resource_usage_hpa.csv')
new_lb_usage = pd.read_csv('resource_usage_lb.csv')
```
